```python

from dotenv import load_dotenv

# .env 파일에서 OPENAI_API_KEY 이름의 key 값을 가져옵니다.
load_dotenv() 
```

```python
from openai import OpenAI

client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        # 상담사 역할 정의
        {
            "role": "system",
            "content": (
                "당신은 전문 심리상담사입니다. "
                "따뜻하고 공감적인 태도로 대화하며, "
                "사용자의 감정을 존중하고 안전한 분위기를 조성하세요. "
                "직접적인 진단이나 치료 지시는 하지 말고, "
                "대화를 통해 사용자가 스스로 생각을 정리할 수 있도록 돕습니다."
            )
        },
        # 내담자(사용자) 메시지
        {
            "role": "user",
            "content": (
                "요즘 일이 너무 많아서 스트레스를 많이 받고 있어요. "
                "밤에도 잠을 잘 못 자고, 계속 불안한 기분이 듭니다."
            )
        }
    ],
    temperature=0.7,  
    # 출력의 창의성과 무작위성을 조절하는 파라미터.(0~1  :사용하는 단어 일반적~창의적)
    max_tokens=300
)
print(response.model_dump_json(indent=2))
# print(response.choices[0].message["content"])
```

```python
print(f' AI : {response.choices[0].message.content}')
```

```python
response = client.responses.create(
    model="gpt-5-nano",
    input="Write a one-sentence bedtime story about a unicorn."
)

print(response.model_dump_json(indent=2))
print(f' AI : {response.output_text}')
```
